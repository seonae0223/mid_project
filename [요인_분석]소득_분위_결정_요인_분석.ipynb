```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns


# íƒ€ê²Ÿ ë³€ìˆ˜ì™€ í”¼ì²˜ ì„¤ì • (ì˜ˆ: Master_Retiredë¥¼ íƒ€ê²Ÿ ë³€ìˆ˜ë¡œ ì„¤ì •)
X = df_income.drop('Income_code', axis=1)
y = df_income['Income_code']

# í•™ìŠµ ë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ë¶„í• 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ëžœë¤ í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ ì´ˆê¸°í™” ë° í•™ìŠµ
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# ì˜ˆì¸¡ ìˆ˜í–‰
y_pred = model.predict(X_test)

# ëª¨ë¸ í‰ê°€
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
roc_auc = roc_auc_score(y_test, model.predict_proba(X_test), multi_class='ovr')

# í‰ê°€ ìˆ˜ì¹˜ ì¶œë ¥
evaluation_metrics = {
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1 Score': f1,
    'ROC AUC': roc_auc
}

evaluation_metrics_df = pd.DataFrame.from_dict(evaluation_metrics, orient='index', columns=['Score'])
print("ëª¨ë¸ í‰ê°€ ìˆ˜ì¹˜:")
print(evaluation_metrics_df)

# í˜¼ë™ í–‰ë ¬ ë° ë¶„ë¥˜ ë³´ê³ ì„œ ì¶œë ¥
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# í”¼ì²˜ ì¤‘ìš”ë„ ì‹œê°í™”
feature_importances = pd.DataFrame(model.feature_importances_, index=X.columns, columns=['importance']).sort_values('importance', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x=feature_importances['importance'], y=feature_importances.index)
plt.title('Feature Importances')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
import seaborn as sns

# íƒ€ê²Ÿ ë³€ìˆ˜ì™€ í”¼ì²˜ ì„¤ì • (ì˜ˆ: 'Income_code'ë¥¼ íƒ€ê²Ÿ ë³€ìˆ˜ë¡œ ì„¤ì •)
X = df_income.drop('Income_code', axis=1)
y = df_income['Income_code']

# í•™ìŠµ ë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ë¶„í• 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# íŒŒì´í”„ë¼ì¸ ì„¤ì •: StandardScalerì™€ RandomForestClassifierë¥¼ í•¨ê»˜ ì‚¬ìš©
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', RandomForestClassifier(random_state=42))
])

# í•˜ì´í¼íŒŒë¼ë¯¸í„° ê·¸ë¦¬ë“œ ì„¤ì •
param_grid = {
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [None, 10, 20, 30],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4]
}

# GridSearchCV ì„¤ì •
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='accuracy')

# GridSearchCVë¡œ ëª¨ë¸ í•™ìŠµ
grid_search.fit(X_train, y_train)

# ìµœì ì˜ ëª¨ë¸ë¡œ ì˜ˆì¸¡ ìˆ˜í–‰
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# ëª¨ë¸ í‰ê°€
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
roc_auc = roc_auc_score(y_test, best_model.predict_proba(X_test), multi_class='ovr')

# í‰ê°€ ìˆ˜ì¹˜ ì¶œë ¥
evaluation_metrics = {
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1 Score': f1,
    'ROC AUC': roc_auc
}

evaluation_metrics_df = pd.DataFrame.from_dict(evaluation_metrics, orient='index', columns=['Score'])
print("ëª¨ë¸ í‰ê°€ ìˆ˜ì¹˜:")
print(evaluation_metrics_df)

# í˜¼ë™ í–‰ë ¬ ìƒì„± ë° ì‹œê°í™”
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Greens')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

# ë¶„ë¥˜ ë³´ê³ ì„œ ì¶œë ¥
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# í”¼ì²˜ ì¤‘ìš”ë„ ì‹œê°í™” ë° ìˆœì„œëŒ€ë¡œ ë‚˜ì—´
feature_importances = pd.DataFrame(best_model.named_steps['model'].feature_importances_, index=X.columns, columns=['importance']).sort_values('importance', ascending=False)

# í”¼ì²˜ ì¤‘ìš”ë„ ìˆœì„œëŒ€ë¡œ ë‚˜ì—´
print("í”¼ì²˜ ì¤‘ìš”ë„ ìˆœì„œëŒ€ë¡œ ë‚˜ì—´:")
print(feature_importances)

# í”¼ì²˜ ì¤‘ìš”ë„ ê·¸ëž˜í”„ ì‹œê°í™”
plt.figure(figsize=(10, 12))
sns.barplot(x=feature_importances['importance'], y=feature_importances.index)
plt.title('Feature Importances in Random Forest Classification')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# ë°ì´í„° ì¤€ë¹„ (ì˜ˆì‹œë¡œ 'Income_code'ë¥¼ íƒ€ê²Ÿ ë³€ìˆ˜ë¡œ ì„¤ì •)
X = df_income.drop('Income_code', axis=1)
y = df_income['Income_code']

# í•™ìŠµ ë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ë¶„í• 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ëžœë¤ í¬ë ˆìŠ¤íŠ¸ íšŒê·€ ëª¨ë¸ ì´ˆê¸°í™”
model = RandomForestRegressor(random_state=42)

# í•˜ì´í¼íŒŒë¼ë¯¸í„° ê·¸ë¦¬ë“œ ì„¤ì •
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# GridSearchCV ì„¤ì • (MSEë¥¼ ê¸°ì¤€ìœ¼ë¡œ ìµœì í™”)
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='neg_mean_squared_error')

# GridSearchCVë¡œ ëª¨ë¸ í•™ìŠµ
grid_search.fit(X_train, y_train)

# ìµœì ì˜ ëª¨ë¸ë¡œ ì˜ˆì¸¡ ìˆ˜í–‰
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# ëª¨ë¸ í‰ê°€
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# í‰ê°€ ìˆ˜ì¹˜ ì¶œë ¥
evaluation_metrics = {
    'Mean Squared Error (MSE)': mse,
    'Mean Absolute Error (MAE)': mae,
    'R^2 Score': r2
}

evaluation_metrics_df = pd.DataFrame.from_dict(evaluation_metrics, orient='index', columns=['Score'])
print("ëª¨ë¸ í‰ê°€ ìˆ˜ì¹˜:")
print(evaluation_metrics_df)

# ì„±ëŠ¥ í‰ê°€ ì§€í‘œì˜ ë§‰ëŒ€ ê·¸ëž˜í”„ ì‹œê°í™”
plt.figure(figsize=(8, 6))
sns.barplot(x=evaluation_metrics_df.index, y='Score', data=evaluation_metrics_df)
plt.title('Performance Metrics')
plt.ylabel('Score')
plt.show()

# í”¼ì²˜ ì¤‘ìš”ë„ ì‹œê°í™”
feature_importances = pd.DataFrame(best_model.feature_importances_, index=X.columns, columns=['importance']).sort_values('importance', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x=feature_importances['importance'], y=feature_importances.index)
plt.title('Feature Importances in Random Forest Regression')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns

# ë°ì´í„° ì¤€ë¹„ (ì˜ˆì‹œë¡œ 'Income_code'ë¥¼ íƒ€ê²Ÿ ë³€ìˆ˜ë¡œ ì„¤ì •)
X = df_income.drop('Income_code', axis=1)
y = df_income['Income_code']

# í•™ìŠµ ë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ë¶„í• 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# íšŒê·€ë¶„ì„ ëª¨ë¸ í•™ìŠµ
logistic_model = Pipeline([
    ('scaler', StandardScaler()),  # íŠ¹ì„± ìŠ¤ì¼€ì¼ë§
    ('classifier', LogisticRegression())
])
logistic_model.fit(X_train, y_train)

# ëžœë¤í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ í•™ìŠµ
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# ì„±ëŠ¥ í‰ê°€ í•¨ìˆ˜
def evaluate_model(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)
    y_pred_prob = model.predict_proba(X_test)[:, 1]

    # í‰ê°€ ì§€í‘œ ê³„ì‚°
    accuracy = metrics.accuracy_score(y_test, y_pred)
    precision = metrics.precision_score(y_test, y_pred, average='weighted')
    recall = metrics.recall_score(y_test, y_pred, average='weighted')
    f1 = metrics.f1_score(y_test, y_pred, average='weighted')
    roc_auc = metrics.roc_auc_score(y_test, model.predict_proba(X_test), multi_class='ovr')

    # ê²°ê³¼ ì¶œë ¥
    print(f'{model_name} Evaluation Metrics:')
    print(f'Accuracy: {accuracy:.4f}')
    print(f'Precision (Weighted): {precision:.4f}')
    print(f'Recall (Weighted): {recall:.4f}')
    print(f'F1 Score (Weighted): {f1:.4f}')
    print(f'ROC-AUC: {roc_auc:.4f}')

    # í˜¼ë™ í–‰ë ¬ ì‹œê°í™”
    cm = metrics.confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", linewidths=.5, cmap='BuGn_r', square=True, cbar=False)

    # Income_codeì˜ ê³ ìœ  ê°’ì— ë§žê²Œ xticks, yticks ì„¤ì •
    labels = sorted(y.unique())  # ê³ ìœ ê°’ì„ ì •ë ¬í•˜ì—¬ ë ˆì´ë¸”ë¡œ ì‚¬ìš©
    plt.xticks(ticks=np.arange(len(labels)) + 0.5, labels=labels, rotation=45)
    plt.yticks(ticks=np.arange(len(labels)) + 0.5, labels=labels, rotation=45)

    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title(f'{model_name} Confusion Matrix')
    plt.show()

    # ROC Curve ì‹œê°í™”
    fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_prob, pos_label=model.classes_[1])
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (area = {roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], color='red', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{model_name} ROC Curve')
    plt.legend(loc="lower right")
    plt.show()

# íšŒê·€ë¶„ì„ ëª¨ë¸ í‰ê°€
evaluate_model(logistic_model, X_test, y_test, "Logistic Regression")

# ëžœë¤í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ í‰ê°€
evaluate_model(rf_model, X_test, y_test, "Random Forest")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# ë°ì´í„° ì¤€ë¹„ (ì˜ˆì‹œë¡œ 'Income_code'ë¥¼ íƒ€ê²Ÿ ë³€ìˆ˜ë¡œ ì„¤ì •)
X = df_income.drop('Income_code', axis=1)
y = df_income['Income_code']

# í•™ìŠµ ë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ë¶„í• 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# GBM ëª¨ë¸ ì´ˆê¸°í™”
gbm = GradientBoostingRegressor(random_state=42)

# í•˜ì´í¼íŒŒë¼ë¯¸í„° ê·¸ë¦¬ë“œ ì„¤ì •
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# GridSearchCV ì„¤ì •
grid_search = GridSearchCV(estimator=gbm, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='r2')

# GridSearchCVë¡œ ëª¨ë¸ í•™ìŠµ
grid_search.fit(X_train, y_train)

# ìµœì ì˜ ëª¨ë¸ë¡œ ì˜ˆì¸¡ ìˆ˜í–‰
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# ëª¨ë¸ í‰ê°€
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# í‰ê°€ ìˆ˜ì¹˜ ì¶œë ¥
evaluation_metrics = {
    'Mean Squared Error (MSE)': mse,
    'Mean Absolute Error (MAE)': mae,
    'R^2 Score': r2
}

evaluation_metrics_df = pd.DataFrame.from_dict(evaluation_metrics, orient='index', columns=['Score'])
print("ëª¨ë¸ í‰ê°€ ìˆ˜ì¹˜:")
print(evaluation_metrics_df)

# í”¼ì²˜ ì¤‘ìš”ë„ ì‹œê°í™”
feature_importances = pd.DataFrame(best_model.feature_importances_, index=X.columns, columns=['importance']).sort_values('importance', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x=feature_importances['importance'], y=feature_importances.index)
plt.title('Feature Importances in GBM')
plt.show()
```
