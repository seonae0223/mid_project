# 랜덤포레스트를 사용한 요인 분석

## 1. 최적의 하이퍼 파라미터 설정 안하고 진행

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns


# 타겟 변수와 피처 설정 (예: Master_Retired를 타겟 변수로 설정)
X = df_income.drop('Income_code', axis=1)
y = df_income['Income_code']

# 학습 데이터와 테스트 데이터로 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 랜덤 포레스트 모델 초기화 및 학습
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# 예측 수행
y_pred = model.predict(X_test)

# 모델 평가
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
roc_auc = roc_auc_score(y_test, model.predict_proba(X_test), multi_class='ovr')

# 평가 수치 출력
evaluation_metrics = {
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1 Score': f1,
    'ROC AUC': roc_auc
}

evaluation_metrics_df = pd.DataFrame.from_dict(evaluation_metrics, orient='index', columns=['Score'])
print("모델 평가 수치:")
print(evaluation_metrics_df)

# 혼동 행렬 및 분류 보고서 출력
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# 피처 중요도 시각화
feature_importances = pd.DataFrame(model.feature_importances_, index=X.columns, columns=['importance']).sort_values('importance', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x=feature_importances['importance'], y=feature_importances.index)
plt.title('Feature Importances')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns

# 타겟 변수와 피처 설정 (예: 'Income_code'를 타겟 변수로 설정)
X = df_income.drop('Income_code', axis=1)
y = df_income['Income_code']

# 학습 데이터와 테스트 데이터로 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 랜덤 포레스트 모델 초기화
model = RandomForestClassifier(random_state=42)

# 하이퍼파라미터 그리드 설정
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# GridSearchCV 설정
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='accuracy')

# GridSearchCV로 모델 학습
grid_search.fit(X_train, y_train)

# 최적의 모델로 예측 수행
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# 모델 평가
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
roc_auc = roc_auc_score(y_test, best_model.predict_proba(X_test), multi_class='ovr')

# 평가 수치 출력
evaluation_metrics = {
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1 Score': f1,
    'ROC AUC': roc_auc
}

evaluation_metrics_df = pd.DataFrame.from_dict(evaluation_metrics, orient='index', columns=['Score'])
print("모델 평가 수치:")
print(evaluation_metrics_df)

# 혼동 행렬 및 분류 보고서 출력
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# 피처 중요도 시각화
feature_importances = pd.DataFrame(best_model.feature_importances_, index=X.columns, columns=['importance']).sort_values('importance', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x=feature_importances['importance'], y=feature_importances.index)
plt.title('Feature Importances')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
import seaborn as sns

# 타겟 변수와 피처 설정 (예: 'Income_code'를 타겟 변수로 설정)
X = df_income.drop('Income_code', axis=1)
y = df_income['Income_code']

# 학습 데이터와 테스트 데이터로 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 파이프라인 설정: StandardScaler와 RandomForestClassifier를 함께 사용
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', RandomForestClassifier(random_state=42))
])

# 하이퍼파라미터 그리드 설정
param_grid = {
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [None, 10, 20, 30],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4]
}

# GridSearchCV 설정
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='accuracy')

# GridSearchCV로 모델 학습
grid_search.fit(X_train, y_train)

# 최적의 모델로 예측 수행
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# 모델 평가
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
roc_auc = roc_auc_score(y_test, best_model.predict_proba(X_test), multi_class='ovr')

# 평가 수치 출력
evaluation_metrics = {
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1 Score': f1,
    'ROC AUC': roc_auc
}

evaluation_metrics_df = pd.DataFrame.from_dict(evaluation_metrics, orient='index', columns=['Score'])
print("모델 평가 수치:")
print(evaluation_metrics_df)

# 혼동 행렬 생성 및 시각화
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Greens')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

# 분류 보고서 출력
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# 피처 중요도 시각화 및 순서대로 나열
feature_importances = pd.DataFrame(best_model.named_steps['model'].feature_importances_, index=X.columns, columns=['importance']).sort_values('importance', ascending=False)

# 피처 중요도 순서대로 나열
print("피처 중요도 순서대로 나열:")
print(feature_importances)

# 피처 중요도 그래프 시각화
plt.figure(figsize=(10, 12))
sns.barplot(x=feature_importances['importance'], y=feature_importances.index)
plt.title('Feature Importances in Random Forest Classification')
plt.show()
```

# 랜덤포레스트 회귀 모델(Random Forest Regression)을 사용한 요인 분석

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 준비 (예시로 'Income_code'를 타겟 변수로 설정)
X = df_income.drop('Income_code', axis=1)
y = df_income['Income_code']

# 학습 데이터와 테스트 데이터로 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 랜덤 포레스트 회귀 모델 초기화
model = RandomForestRegressor(random_state=42)

# 하이퍼파라미터 그리드 설정
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# GridSearchCV 설정 (MSE를 기준으로 최적화)
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='neg_mean_squared_error')

# GridSearchCV로 모델 학습
grid_search.fit(X_train, y_train)

# 최적의 모델로 예측 수행
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# 모델 평가
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# 평가 수치 출력
evaluation_metrics = {
    'Mean Squared Error (MSE)': mse,
    'Mean Absolute Error (MAE)': mae,
    'R^2 Score': r2
}

evaluation_metrics_df = pd.DataFrame.from_dict(evaluation_metrics, orient='index', columns=['Score'])
print("모델 평가 수치:")
print(evaluation_metrics_df)

# 성능 평가 지표의 막대 그래프 시각화
plt.figure(figsize=(8, 6))
sns.barplot(x=evaluation_metrics_df.index, y='Score', data=evaluation_metrics_df)
plt.title('Performance Metrics')
plt.ylabel('Score')
plt.show()

# 피처 중요도 시각화
feature_importances = pd.DataFrame(best_model.feature_importances_, index=X.columns, columns=['importance']).sort_values('importance', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x=feature_importances['importance'], y=feature_importances.index)
plt.title('Feature Importances in Random Forest Regression')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 준비 (예시로 'Income_code'를 타겟 변수로 설정)
X = df_income.drop('Income_code', axis=1)
y = df_income['Income_code']

# 학습 데이터와 테스트 데이터로 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 회귀분석 모델 학습
logistic_model = Pipeline([
    ('scaler', StandardScaler()),  # 특성 스케일링
    ('classifier', LogisticRegression())
])
logistic_model.fit(X_train, y_train)

# 랜덤포레스트 모델 학습
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# 성능 평가 함수
def evaluate_model(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)
    y_pred_prob = model.predict_proba(X_test)[:, 1]

    # 평가 지표 계산
    accuracy = metrics.accuracy_score(y_test, y_pred)
    precision = metrics.precision_score(y_test, y_pred, average='weighted')
    recall = metrics.recall_score(y_test, y_pred, average='weighted')
    f1 = metrics.f1_score(y_test, y_pred, average='weighted')
    roc_auc = metrics.roc_auc_score(y_test, model.predict_proba(X_test), multi_class='ovr')

    # 결과 출력
    print(f'{model_name} Evaluation Metrics:')
    print(f'Accuracy: {accuracy:.4f}')
    print(f'Precision (Weighted): {precision:.4f}')
    print(f'Recall (Weighted): {recall:.4f}')
    print(f'F1 Score (Weighted): {f1:.4f}')
    print(f'ROC-AUC: {roc_auc:.4f}')

    # 혼동 행렬 시각화
    cm = metrics.confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", linewidths=.5, cmap='BuGn_r', square=True, cbar=False)

    # Income_code의 고유 값에 맞게 xticks, yticks 설정
    labels = sorted(y.unique())  # 고유값을 정렬하여 레이블로 사용
    plt.xticks(ticks=np.arange(len(labels)) + 0.5, labels=labels, rotation=45)
    plt.yticks(ticks=np.arange(len(labels)) + 0.5, labels=labels, rotation=45)

    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title(f'{model_name} Confusion Matrix')
    plt.show()

    # ROC Curve 시각화
    fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_prob, pos_label=model.classes_[1])
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (area = {roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], color='red', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{model_name} ROC Curve')
    plt.legend(loc="lower right")
    plt.show()

# 회귀분석 모델 평가
evaluate_model(logistic_model, X_test, y_test, "Logistic Regression")

# 랜덤포레스트 모델 평가
evaluate_model(rf_model, X_test, y_test, "Random Forest")
```

## GBM 을 활용한 소득 분위 결정 요인 분석

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 준비 (예시로 'Income_code'를 타겟 변수로 설정)
X = df_income.drop('Income_code', axis=1)
y = df_income['Income_code']

# 학습 데이터와 테스트 데이터로 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# GBM 모델 초기화
gbm = GradientBoostingRegressor(random_state=42)

# 하이퍼파라미터 그리드 설정
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# GridSearchCV 설정
grid_search = GridSearchCV(estimator=gbm, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='r2')

# GridSearchCV로 모델 학습
grid_search.fit(X_train, y_train)

# 최적의 모델로 예측 수행
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# 모델 평가
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# 평가 수치 출력
evaluation_metrics = {
    'Mean Squared Error (MSE)': mse,
    'Mean Absolute Error (MAE)': mae,
    'R^2 Score': r2
}

evaluation_metrics_df = pd.DataFrame.from_dict(evaluation_metrics, orient='index', columns=['Score'])
print("모델 평가 수치:")
print(evaluation_metrics_df)

# 피처 중요도 시각화
feature_importances = pd.DataFrame(best_model.feature_importances_, index=X.columns, columns=['importance']).sort_values('importance', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x=feature_importances['importance'], y=feature_importances.index)
plt.title('Feature Importances in GBM')
plt.show()
```
